1. Write a Python function that takes a list and returns a new list with distinct elements from the
first list.
Sample List : [1,2,3,3,3,3,4,5]
Unique List : [1, 2, 3, 4, 5]

```python
def unique_list(lst):
    seen = set()
    result = []
    for item in lst:
        if item not in seen:
            seen.add(item)
            result.append(item)
    return result

sample = [1, 2, 3, 3, 3, 3, 4, 5]
print("Sample List:", sample)
print("unique_list", unique_list(sample))
```

2. Write a Python function that takes a number as a parameter and checks whether the number
is prime or not.

```python
def is_prime(n):
    if n < 2:
        return False
    for i in range(2, int(n ** 0.5) + 1):
        if n % i == 0:
            return False
    return True

print(is_prime(7))
print(is_prime(10))
print(is_prime(2))
```

4. A positive integer is called a perfect number if it is equal to the sum of all of its divisors,
including 1 but excluding the number itself. For example, 6 = 1 + 2 + 3. Write a
function sumOfDivisors that takes a positive integer n and returns the sum of all its divisors
(excluding n).

```python
def sumOfDivisors(n):
    total = 0
    for i in range(1, n):
        if n % i == 0:
            total += i
    return total
print(sumOfDivisors(6))
print(sumOfDivisors(28))
print(sumOfDivisors(10))
```

5. The smallest perfect number is 6. Write a program that finds and prints out the next perfect number. Hint:
You should be able your sumOfDiviors function from your solution to previous question.

```python
def sumOfDivisors(n):
    total = 0
    for i in range(1, n):
        if n % i == 0:
            total += i
    return total

num = 7  # start checking after 6

while True:
    if sumOfDivisors(num) == num:
        print("Next perfect number is:", num)
        break
    num += 1
```

6. Write the python program to find the mean median from the following list:
Marks = [20, 30, 35, 55, 70, 80, 90] Do not use the library function if there is one

```python
marks = [20, 30, 35, 55, 70, 80, 90]

# Mean
total = 0
for m in marks:
    total += m

mean = total / len(marks)

# Median
n = len(marks)

# (List is already sorted, but normally you would sort it)
if n % 2 == 1:
    median = marks[n // 2]
else:
    mid1 = marks[n // 2]
    mid2 = marks[(n // 2) - 1]
    median = (mid1 + mid2) / 2

# Output
print("Mean =", mean)
print("Median =", median)
```

7. Write a function that converts a decimal number to binary number

```python
def decimal_to_binary(n):
    if n == 0:
        return "0"

    binary = ""

    while n > 0:
        remainder = n % 2
        binary = str(remainder) + binary
        n = n // 2

    return binary

num = 10
print("Binary of", num, "is:", decimal_to_binary(num))
```

8. Write a function prodDigits() that inputs a number and returns the product of digits of that
number.

```python
def prodDigits(n):
    product = 1

    while n > 0:
        digit = n % 10
        product *= digit
        n = n // 10

    return product
num = 1234
print("Product of digits:", prodDigits(num))
```
