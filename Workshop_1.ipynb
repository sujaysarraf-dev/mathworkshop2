```python
def order_numbers(number1, number2):
    if number2 > number1:
        return number1, number2
    else:
        return number2, number1
smaller, bigger = order_numbers(100, 99)
print(smaller, bigger)
```

```python
def myPow(x, n):
    result = 1
    count = 0
    while count < n:
       result = result * x
       count = count + 1
    return result
print(myPow(2, 3))
```

```python
def SumDigits(n):
    total = 0
    while n > 0:
        digit = n % 10
        total = total + digit
        n = n // 10
    return total
print(SumDigits(234))
```

```python
def squaredSum(numbers):
    total = 0
    for num in numbers:
        total = total + (num * num)
    return total
print(squaredSum([2, 3, 4]))
```

```python
def sumDigits(num):
    total = 0
    while num > 0:
        digit = num % 10
        total = total + digit
        num = num // 10
    return total
print(sumDigits(123456))
```
